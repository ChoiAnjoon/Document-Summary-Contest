```python
import pandas as pd
import numpy as np
import json
```

```python
with open('/content/drive/MyDrive/신문기사 요약/test.jsonl', "r", encoding="utf-8") as f:
    jsonl = list(f)

test_data = []
for json_str in jsonl:
    test_data.append(json.loads(json_str))
    
submission = pd.read_csv('/content/drive/MyDrive/신문기사 요약/sample_submission.csv')
```

```python
submission.head()
```

```python
test_data[0]
```

```python
for data in test_data:
    did = int(data['id'])
    summary = data['article_original'][0] + data['article_original'][-1]
    idx = submission[submission['id']==did].index
    submission.loc[idx, 'summary'] = summary
```
